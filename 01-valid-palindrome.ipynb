### 01. Valid Palindrome

The problem is from [LeetCode 125. Valid Palindrome](https://leetcode.com/problems/valid-palindrome/).

**Problem** Given a string, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

**Note** For the purpose of this problem, we define empty string as valid palindrome.

**Example 1:**
~~~~
 Input: "A man, a plan, a canal: Panama 
 Output: true
~~~~

**Example 2:**
~~~~
 Input: "race a car"
 Output: false
~~~~

**Constraints:**
* `s` consists only of printable ASCII characters.

```python
## Python 3
import re
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        sentence = re.sub('[^a-zA-Z0-9]','',s)
        return sentence == sentence[::-1]
```

### 풀이방법

1. 정규표현식
    * re.sub(pattern, repl, string, [count], [flags]) 함수는 string 내에 있는 pattern을 repl로 치환해준다.
    * [^a-zA-Z0-9] 문자 및 숫자를 제외한 어떤 문자를 의미한다.
2. 슬라이싱
    * [::-1]은 배열을 역순으로 표현해준다.
